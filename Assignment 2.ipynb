```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
```

```python
df
```

```python
df.info()
```

```python
x = df["total"].values
y = df["alcohol"].values
z = df["speeding"].values
plt.figure()
plt.plot(x, label='Total Car Crash', color='blue')
plt.plot(y, label='Alcohol related', color='green')
plt.plot(z, label='Speed related', color='red')
plt.legend()
sns.set()
plt.show()
```
